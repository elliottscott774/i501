```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def plot_binary_array(array):
    plt.imshow(array, cmap='binary')
    plt.axis('off')
    plt.show()
```

```python
def game_of_life_step(array):
    neighbors = np.zeros(array.shape, dtype=int)
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue
            neighbors += np.roll(np.roll(array, i, axis=0), j, axis=1)
    new_array = np.where((array == 1) & ((neighbors == 2) | (neighbors == 3)), 1, 0)
    new_array += np.where((array == 0) & (neighbors == 3), 1, 0)
    return new_array
```

```python
def run_game_of_life(initial_array, steps=10):
    array = initial_array.copy()
    for _ in range(steps):
        plot_binary_array(array)
        array = game_of_life_step(array)
        time.sleep(1)  # Pause for 1 second between steps
```

```python
# Create a random binary array
np.random.seed(42)  # Set a seed for reproducibility
initial_array = np.random.randint(2, size=(10, 10))

# Run the Game of Life
run_game_of_life(initial_array, steps=10)
```
